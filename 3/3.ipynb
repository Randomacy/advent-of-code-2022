```python
with open("input.txt") as f:
    lines = f.readlines()
```

```python
score = dict(zip(list(string.ascii_letters), list(range(1, 53))))
score
```

```python
priority_score  = 0
for line in lines:
    sack_1 = line[:int(len(line)/2)]
    sack_2 = line[int(len(line)/2):]
    common = list(set(sack_1).intersection(set(sack_2)))[0]
    priority_score += score[common]
```

```python
group_of_elves = []
group = []
for line in lines:
    if len(group_of_elves) != 99:
        if len(group) != 3:
            group.append(line)
        else:
            group_of_elves.append(group)
            group = []
            group.append(line)
    else:
        group.append(line)
        group_of_elves.append(group)
```
