### Part I

### Define helper functions

```python
def parse_instruction(instruction):
    split = instruction.split(" ")
    return int(split[1]), int(split[3]), int(split[5])
```

```python
# test the parser
for i in moves:
    print(parse_instruction(i))
```

```python
# define what happens to stack when instructions are given
def move_stack(stack, instructions):
    num_boxes = instructions[0]
    first_stack = instructions[1] - 1
    end_stack = instructions[2] - 1
    
    for num in range(num_boxes):
        removed = stack[first_stack].pop()
        stack[end_stack].append(removed)
```

```python
for i in moves:
    move_stack(stack_of_stacks, parse_instruction(i))
```

```python
# final stack
stack_of_stacks
```

```python
# get the items on top of the stack
for i in stack_of_stacks:
    print(i[-1])
```

### Part II

```python
# decided to just manually type out the stacks because I'm lazy
stack_1 = ['Z', 'T', 'F', 'R', 'W', 'J', 'G']
stack_2 = ['G', 'W', 'M']
stack_3 = ['J', 'N', 'H', 'G']
stack_4 = ['J', 'R', 'C', 'N', 'W']
stack_5 = ['W', 'F', 'S', 'B', 'G', 'Q', 'V', 'M']
stack_6 = ['S', 'R', 'T', 'D', 'V', 'W', 'C']
stack_7 = ['H', 'B', 'N', 'C', 'D', 'Z', 'G', 'V']
stack_8 = ['S', 'J', 'N', 'M', 'G', 'C']
stack_9 = ['G', 'P', 'N', 'W', 'C', 'J', 'D', 'L']

# declare stack of stacks
stack_of_stacks = [stack_1, stack_2, stack_3, 
                   stack_4, stack_5, stack_6, 
                   stack_7, stack_8, stack_9]
```

```python
# define what happens to stack when instructions are given
def move_stack_v2(stack, instructions):
    num_boxes = instructions[0]
    first_stack = instructions[1] - 1
    end_stack = instructions[2] - 1
    
    removed = stack[first_stack][-num_boxes:]
    stack[first_stack] = stack[first_stack][:-num_boxes]
    stack[end_stack].extend(removed)
```

```python
for i in moves:
    move_stack_v2(stack_of_stacks, parse_instruction(i))
```
