```python
with open("input.txt") as f:
    lines = f.readlines()
```

```python
lines
```

```python
new_lines = []

for i in lines:
    j = i.strip()
    j = j.split(" ")
    new_lines.append(j)
    
new_lines
```

```python
move_dict = {"X": 1, "Y": 2, "Z": 3}
score = 0

for f, s in new_lines:
    # win condition
    if (f == 'A' and s == "Y") or (f == 'B' and s == "Z") or (f == 'C' and s == 'X'):
        score += 6
        score += move_dict[s]
    # lose condition
    elif (f == 'A' and s == "Z") or (f == 'B' and s == "X") or (f == 'C' and s == 'Y'):
        score += move_dict[s]
    else:
        score += 3
        score += move_dict[s]
```

```python
score
```

```python
# X = lose
# Y = draw
# Z = win

move_dict = {"X": 1, "Y": 2, "Z": 3}
score_dict = {"A": 1, "B": 2, "C": 3}
win_dict = {"A": "B", "B": "C", "C": "A"}
lose_dict = {"A": "C", "B": "A", "C": "B"}
draw_dict = {"A": "A", "B": "B", "C": "C"}

score = 0

for f, s in new_lines:
    # assess condition
    if s == 'Z':
        score += score_dict[win_dict[f]]
        score += 6
        
    elif s == 'Y':
        score += score_dict[draw_dict[f]]
        score += 3
        
    else:
        score += score_dict[lose_dict[f]]
```
