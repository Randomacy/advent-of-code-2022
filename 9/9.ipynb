```python
class Rope:
    def __init__(self, head_x, head_y, tail_x, tail_y):
        self.head_x = head_x
        self.head_y = head_y
        self.tail_x = tail_x
        self.tail_y = tail_y
        self.steps = 0
        self.positions = set()

    def move_head(self, direction, steps):
        for step in range(steps):
            if direction == 'R':
                if step == 0:
                    self.head_x += 1
                else:
                    self.head_x += 1
            elif direction == 'U':
                if step == 0:
                    self.head_y += 1
                else:
                    self.head_y += 1
            elif direction == 'L':
                if step == 0:
                    self.head_x -= 1
                else:
                    self.head_x -= 1
            else:
                if step == 0:
                    self.head_y -= 1
                else:
                    self.head_y -= 1

    def get_head_pos(self):
        return (self.head_x, self.head_y)
    def get_tail_pos(self):
        return (self.tail_x, self.tail_y)
```

```python
class Rope:
    def __init__(self, head_x, head_y, tail_x, tail_y):
        self.head_x = head_x
        self.head_y = head_y
        self.tail_x = tail_x
        self.tail_y = tail_y
        self.steps_taken = 0
        self.positions = set()

    def move_head(self, direction, steps):
        for step in range(steps):
            if direction == 'R':
                if step == 0:
                    self.head_x += 1
                    self.positions.add((self.tail_x, self.tail_y))
                else:
                    if ((self.head_x, self.head_y) == (self.tail_x, self.tail_y)):
                        self.head_x += 1
                        self.positions.add((self.tail_x, self.tail_y))
                        break
                    else:
                        self.tail_x = self.head_x
                        self.tail_y = self.head_y
                        self.positions.add((self.tail_x, self.tail_y))
                    self.head_x += 1
                    self.positions.add((self.tail_x, self.tail_y))
            elif direction == 'U':
                if step == 0:
                    self.head_y += 1
                    self.positions.add((self.tail_x, self.tail_y))
                else:
                    if ((self.head_x, self.head_y) == (self.tail_x, self.tail_y)):
                        self.head_y += 1
                        self.positions.add((self.tail_x, self.tail_y))
                        break
                    else:
                        self.tail_x = self.head_x
                        self.tail_y = self.head_y
                        self.positions.add((self.tail_x, self.tail_y))
                    self.head_y += 1
                    self.positions.add((self.tail_x, self.tail_y))
            elif direction == 'L':
                if step == 0:
                    self.head_x -= 1
                    self.positions.add((self.tail_x, self.tail_y))
                else:
                    if ((self.head_x, self.head_y) == (self.tail_x, self.tail_y)):
                        self.head_x += 1
                        self.positions.add((self.tail_x, self.tail_y))
                        break
                    else:
                        self.tail_x = self.head_x
                        self.tail_y = self.head_y
                        self.positions.add((self.tail_x, self.tail_y))
                    self.head_x -= 1
                    self.positions.add((self.tail_x, self.tail_y))
            else:
                if step == 0:
                    self.head_y -= 1
                    self.positions.add((self.tail_x, self.tail_y))
                else:
                    if ((self.head_x, self.head_y) == (self.tail_x, self.tail_y)):
                        self.head_y += 1
                        self.positions.add((self.tail_x, self.tail_y))
                        break
                    else:
                        self.tail_x = self.head_x
                        self.tail_y = self.head_y
                        self.positions.add((self.tail_x, self.tail_y))
                    self.head_y -= 1
                    self.positions.add((self.tail_x, self.tail_y))

    def get_head_pos(self):
        return (self.head_x, self.head_y)
    def get_tail_pos(self):
        return (self.tail_x, self.tail_y)
```
