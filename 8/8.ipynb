### Define helper array conversion function

```python
# process list of text into proper array
def process_array(list_of_text):
    array = []

    for line in list_of_text:
        temp = list(line)
        temp = [int(i) for i in temp]
        array.append(temp)  
    return array
```

```python
# check it out
processed_lines = process_array(lines)
print(processed_lines)
```

```python
# do the same for test lines
processed_test_lines = process_array(test_lines)
print(processed_test_lines)
```

### Define perimeter length calculator

```python
def find_perimeter_len(processed_lines):
    top = len(processed_lines[0])
    bottom = len(processed_lines[-1])
    left_and_right = len(processed_lines[1:-1]) * 2
    
    return top + bottom + left_and_right
```

### Part II
Now we need to get the view score from any tree in the grid. Let's start with a helper function to get the view score.

```python
# define function to output a score based on the current tree and the direction's range
def view_score(current, tree_range):
    score = 0
    for tree in tree_range:
        if current > tree:
            score += 1
        else:
            score += 1
            break
    return score
```

```python
# test the function against the test data
print(view_score(5, [3]))
print(view_score(5, [1, 2]))
print(view_score(5, [3, 5, 3]))
print(view_score(5, [5, 2]))
```

```python
# putting it together for the test data
visible = 0
max_score = 0

for i in range(1, len(processed_test_lines) - 1):
    for j in range(1, len(processed_test_lines[i]) - 1):
        current = processed_test_lines[i][j]
        top_range = [processed_test_lines[a][j] for a in range(i)][::-1]
        right_range = [processed_test_lines[i][b] for b in range(j+1, len(processed_test_lines[i]))]
        bottom_range = [processed_test_lines[c][j] for c in range(i+1, len(processed_test_lines))]
        left_range = [processed_test_lines[i][d] for d in range(j)][::-1]
        
        top_view = view_score(current, top_range)
        right_view = view_score(current, right_range)
        bottom_view = view_score(current, bottom_range)
        left_view = view_score(current, left_range)
        
        scenic_score = top_view * right_view * bottom_view * left_view
        
        if scenic_score > max_score:
            max_score = scenic_score

max_score
```
